```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_olist_orders = pd.read_csv('https://raw.githubusercontent.com/GlayconC/AnaliseDeDados/main/olist_orders_dataset.csv')
```

```python
df_olist_orders.head()
```

```python
column_translations = {
    'order_id': 'id_pedido',
    'customer_id': 'id_cliente',
    'order_status': 'status_pedido',
    'order_purchase_timestamp': 'data_hora_compra',
    'order_approved_at': 'data_hora_aprovacao',
    'order_delivered_carrier_date': 'data_entrega_transportadora',
    'order_delivered_customer_date': 'data_entrega_cliente',
    'order_estimated_delivery_date': 'data_entrega_estimada'
}

df_olist_orders = df_olist_orders.rename(columns=column_translations)
display(df_olist_orders.head())
```

```python
df_olist_orders['data_entrega_cliente'] = pd.to_datetime(df_olist_orders['data_entrega_cliente'])
```

```python
df_olist_orders.dtypes
```

```python
df_olist_orders['data_hora_compra'] = pd.to_datetime(df_olist_orders['data_hora_compra'])
df_olist_orders['data_hora_aprovacao'] = pd.to_datetime(df_olist_orders['data_hora_aprovacao'])
df_olist_orders['data_entrega_transportadora'] = pd.to_datetime(df_olist_orders['data_entrega_transportadora'])
df_olist_orders['data_entrega_estimada'] = pd.to_datetime(df_olist_orders['data_entrega_estimada'])
```

```python
# Remover Colunas com Nulos

df_orders_limpo = df_olist_orders.dropna()
```

```python
# Tempo m√©dio para aprovar
tempo_medio = (df_orders_limpo['data_hora_aprovacao'] - df_orders_limpo['data_hora_compra']).mean()
tempo_medio_arredondado_aprovar = pd.Timedelta(seconds=int(tempo_medio.total_seconds()))

print(f"Tempo m√©dio: {tempo_medio_arredondado_aprovar}")
```

```python
# Tempo m√©dio para entregar a transportadora ap√≥s aprovar


tempo_medio = (df_orders_limpo['data_entrega_transportadora'] - df_orders_limpo['data_hora_aprovacao']).mean()
tempo_medio_arredondado_entregar_transportadora = pd.Timedelta(seconds=int(tempo_medio.total_seconds()))

print(f"Tempo m√©dio: {tempo_medio_arredondado_entregar_transportadora}")
```

```python
# Tempo m√©dio para a transportadora entregar aos clientes

tempo_medio = (df_orders_limpo['data_entrega_cliente'] - df_orders_limpo['data_entrega_transportadora']).mean()
tempo_medio_arredondado_transportadora_clientes = pd.Timedelta(seconds=int(tempo_medio.total_seconds()))

print(f"Tempo m√©dio: {tempo_medio_arredondado_transportadora_clientes}")
```

```python
# Tempo da Compra At√© Entregar ao Cliente

tempo_medio = (df_orders_limpo['data_entrega_cliente'] - df_orders_limpo['data_hora_compra']).mean()
tempo_medio_arredondado_compra_clientes = pd.Timedelta(seconds=int(tempo_medio.total_seconds()))

print(f"Tempo m√©dio: {tempo_medio_arredondado_compra_clientes}")
```

```python
print("=" * 60)
print("üì¶ AN√ÅLISE COMPLETA: ATRASOS NAS ENTREGAS")
print("=" * 60)

# 1. Calcular atraso
tempo_atraso = (df_orders_limpo['data_entrega_cliente'] -
                df_orders_limpo['data_entrega_estimada'])

# 2. Categorizar
atrasos = tempo_atraso[tempo_atraso > pd.Timedelta(0)]
adiantados = tempo_atraso[tempo_atraso < pd.Timedelta(0)]
no_prazo = tempo_atraso[tempo_atraso == pd.Timedelta(0)]

# 3. Estat√≠sticas
total = len(tempo_atraso)
print(f"\nüìä RESUMO:")
print(f"‚îú‚îÄ Total de pedidos: {total}")
print(f"‚îú‚îÄ Atrasados: {len(atrasos)} ({len(atrasos)/total*100:.1f}%)")
print(f"‚îú‚îÄ Adiantados: {len(adiantados)} ({len(adiantados)/total*100:.1f}%)")
print(f"‚îî‚îÄ No prazo: {len(no_prazo)} ({len(no_prazo)/total*100:.1f}%)")

print(f"\n‚è±Ô∏è  ATRASOS:")
print(f"‚îú‚îÄ M√©dia: {atrasos.mean().days:.1f} dias")
print(f"‚îú‚îÄ Mediana: {atrasos.median().days} dias")
print(f"‚îú‚îÄ M√°ximo: {atrasos.max().days} dias")
print(f"‚îî‚îÄ M√≠nimo: {atrasos.min().days} dias")

# 4. Visualiza√ß√µes
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histograma
atrasos_dias = atrasos.dt.days
axes[0].hist(atrasos_dias, bins=50, color='coral', edgecolor='black', alpha=0.7)
axes[0].axvline(atrasos_dias.mean(), color='red', linestyle='--', linewidth=2,
                label=f'M√©dia: {atrasos_dias.mean():.1f} dias')
axes[0].set_title('Distribui√ß√£o dos Atrasos', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Dias de Atraso')
axes[0].set_ylabel('Frequ√™ncia')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Pizza
categorias = ['Atrasados', 'Adiantados', 'No Prazo']
valores = [len(atrasos), len(adiantados), len(no_prazo)]
cores = ['#ff6b6b', '#51cf66', '#74c0fc']
axes[1].pie(valores, labels=categorias, autopct='%1.1f%%',
            colors=cores, startangle=90)
axes[1].set_title('Propor√ß√£o de Entregas', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Defina o ano e o m√™s desejados
ano1 = 2017
mes1 = 12 # Dezembro

ano2 = 2016
mes2 = 12  # Dezembro

# Cria as datas de in√≠cio e fim do m√™s
data_inicio_mes1 = pd.to_datetime(f'{ano1}-{mes1}-01')
data_inicio_mes2 = pd.to_datetime(f'{ano2}-{mes2}-01')

# Verificar o pr√≥ximo m√™s
proximo_mes1 = mes1 + 1
proximo_ano1 = ano1
proximo_mes2 = mes2 + 1
proximo_ano2 = ano2

# Se o m√™s for dezembro, o pr√≥ximo m√™s √© janeiro do pr√≥ximo ano
if mes1 == 12:
    proximo_mes1 = 1
    proximo_ano1 = ano1 + 1
if mes2 == 12:
    proximo_mes2 = 1
    proximo_ano2 = ano1 + 1

data_fim_mes1= pd.to_datetime(f'{proximo_ano1}-{proximo_mes1}-01') - pd.Timedelta(days=1)
data_fim_mes2= pd.to_datetime(f'{proximo_ano2}-{proximo_mes2}-01') - pd.Timedelta(days=1)

# Filtre o DataFrame para o m√™s espec√≠fico
vendas_mes1 = df_orders_limpo[(df_orders_limpo['data_hora_compra'] >= data_inicio_mes1) &
                               (df_orders_limpo['data_hora_compra'] <= data_fim_mes1)]

vendas_mes2 = df_orders_limpo[(df_orders_limpo['data_hora_compra'] >= data_inicio_mes2) &
                               (df_orders_limpo['data_hora_compra'] <= data_fim_mes2)]


print(f"Total de vendas em {data_inicio_mes1.strftime('%B de %Y')}: {len(vendas_mes1)}")
print(f"Total de vendas em {data_inicio_mes2.strftime('%B de %Y')}: {len(vendas_mes2)}")
```

```python
ano = 2017
meses = []
vendas = []

for mes in range(1, 13):
    data_inicio_mes = pd.to_datetime(f'{ano}-{mes}-01')

    if mes == 12:
        data_fim_mes = pd.to_datetime(f'{ano+1}-01-01') - pd.Timedelta(days=1)
    else:
        data_fim_mes = pd.to_datetime(f'{ano}-{mes+1}-01') - pd.Timedelta(days=1)

    vendas_mes = df_orders_limpo[(df_orders_limpo['data_hora_compra'] >= data_inicio_mes) &
                                 (df_orders_limpo['data_hora_compra'] <= data_fim_mes)]

    meses.append(mes)
    vendas.append(len(vendas_mes))

# Criar gr√°fico
plt.figure(figsize=(14, 6))
plt.plot(meses, vendas, marker='o', linewidth=2.5, markersize=8, color='#004E89')

# Adicionar valores em cima dos pontos
for i, (x, y) in enumerate(zip(meses, vendas)):
    plt.text(x, y + max(vendas)*0.02, f'{y:,}', ha='center',
             fontsize=10, fontweight='bold')

# Configura√ß√µes
plt.title(f'Evolu√ß√£o de Vendas Mensais - {ano}', fontsize=14, fontweight='bold')
plt.xlabel('M√™s', fontsize=12)
plt.ylabel('N√∫mero de Pedidos', fontsize=12)
plt.xticks(meses, ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
                    'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
plt.grid(True, alpha=0.3, linestyle='--')
plt.ylim(0, max(vendas) * 1.15)
plt.tight_layout()
plt.show()

print(f"\nüìä Resumo {ano}:")
print(f"Maior venda: {max(vendas):,} pedidos (m√™s {vendas.index(max(vendas)) + 1})")
print(f"Menor venda: {min(vendas):,} pedidos (m√™s {vendas.index(min(vendas)) + 1})")
print(f"M√©dia mensal: {sum(vendas)/len(vendas):,.0f} pedidos")
```
